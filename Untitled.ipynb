```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hd = pd.read_csv('heart_data.csv')
hd
```

```python
hd.head(10)
```

```python
hd.info()
```

```python
hd['age'] = hd['age']//365
```

```python
hd
```

```python
hd['gender'] = hd['gender'].map({2:0,1:1})
```

```python
hd
```

```python
hd.drop(['index','id'], axis=1,inplace= True)
hd.head(10)
```

```python
sns.histplot(hd['age'],kde=True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(hd['weight'],kde=True)
plt.show()
```

```python
hd.describe()
```

```python
sns.jointplot(x='height', y='weight', data = hd, kind = 'kde', color = 'red')
```

```python
sns.barplot(x='gender', y='cholesterol', data = hd)
```

```python
chol = hd['cholesterol'].value_counts()
sns.barplot(x=chol.index, y=chol.values)
plt.xlabel('cholesterol level')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='cholesterol',hue='cardio',data= hd)
plt.title('Cholesterol')
plt.show()
```

```python
plt.figure(figsize=(10,10))
dataplot = sns.heatmap(hd.corr(), annot=True)
dataplot
```
